```python
import csv
import random
from faker import Faker
from datetime import date

fake=Faker()

credit_terms_exact = [
    "30 Days Net", "7 Days Net", "Immediate", "45 Days Net",
    "21 Days Net", "60 Days Net", "14 Days Net"
]

# Generate 5000 unique customer entries with unique customer IDs
unique_customer_rows = []
used_ids = set()

default_email='delinchackrapurkal@gmail.com'

start_date = date(2014, 1, 1)
end_date = date(2024, 12, 31)


while len(unique_customer_rows) < 5000:
    cust_id = random.randint(1000, 99999)
    if cust_id in used_ids:
        continue
    used_ids.add(cust_id)
    name = fake.company()
    terms = random.choice(credit_terms_exact)
    reg_date = fake.date_between(start_date, end_date)
    reg_date_str = reg_date.strftime("%Y-%m-%d")

    unique_customer_rows.append([cust_id, name, terms, default_email,reg_date_str])

# Write to CSV
unique_file_path = "customer_master.csv"
with open(unique_file_path, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Customer_No", "Customer_Name", "Terms", "email", "Entered_date" ])
    writer.writerows(unique_customer_rows)

unique_file_path
```

```python
import csv
import random
from faker import Faker
from datetime import date

fake=Faker()


# Generate 5000 unique customer entries with unique customer IDs
unique_invoice = []
used_ids = set()
credit_terms_exact = [
    "30 Days Net", "7 Days Net", "Immediate", "45 Days Net",
    "21 Days Net", "60 Days Net", "14 Days Net"
]

start_date = date(2024, 5, 1)
end_date = date(2024, 12, 31)


while len(unique_invoice) < 100000:
    invoice_id = random.randint(1000, 999999)
    
    if invoice_id in used_ids:
        continue
    used_ids.add(invoice_id)
    invoice_date = fake.date_between(start_date, end_date)
    value = round(random.uniform(0.2, 1.0) ** 0.5 * 1_000_000, 2)
    customer_name = fake.company()
    credit_terms = random.choice(credit_terms_exact)
    invoice_gross_value=max(value, 200_000)
    invoice_currency=random.choice(["KES","USD"])
    email=fake.email()
    



    

    unique_invoice.append([invoice_id,invoice_date,invoice_gross_value,invoice_currency,customer_name, credit_terms ,email])

# Write to CSV
unique_file_path = "customer_invoice.csv"
with open(unique_file_path, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["invoice_id", "invoice_date", "invoice_gross_value", "invoice_currency","customer_name","credit_terms","email"  ])
    writer.writerows(unique_invoice)

unique_file_path
```
